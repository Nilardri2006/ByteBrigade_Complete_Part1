Operators in Python:

5. Identity Operators

1. Arithmetic Operators

```python
num_1 = 10
num_2 = 20

# addition
sum = num_1 + num_2
print("Sum = ", sum)

# subtraction
diff = num_1 - num_2
print("Difference = ", diff)

# multiplication
prod = num_1 * num_2
print("Product = ", prod)

# division
div = num_1 / num_2
print("Division = ", div)

# exponent
exp = num_1 ** num_2
print("Exponent = ", exp)

# modulus
mod = num_1 % num_2
print("Modulus = ", mod)
```

3. Comparison Operators

```python
a = 5
b = 10

print( a==b)
print( a!=b)
print( a>b)
print( a<b)
print( a>=b)
print( a<=b)
```

4. Logical Operators


and

or

not

```python
a = 10

print(a>20 and a>5)
print(a>20 or a>5)
print(not(a>8 and a>5))
```

5. Identity Operator

is

is not

```python
x = 5
y = 5
print(x is y)
```

```python
x = 5
y = 6
print(x is y)
```

```python
x = 5
y = 5
print(x is not y)
```

```python
x = 5
y = 10
print(x is not y)
```

6. Membership Operator

in

not in

```python
a = 5
b = 10

c = [1, 2, 3, 4, 5]
print(a in c)
print(b in c)
```

```python
a = 5
b = 10

c = [1, 2, 3, 4, 5]
print(a not in c)
print(b not in c)
```
